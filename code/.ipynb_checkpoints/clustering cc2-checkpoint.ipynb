```python
# Functions for EDA
def remove_duplicates(df):
    if(any(df.duplicated())):
        return df.drop_duplicates(keep='first')
    
    print('No duplicates')
    return df

def count_nulls(df):
    # Count the number of rows with null values
    num_rows_with_nulls = df.isnull().any(axis=1).sum()
    
    # Count the number of null values in each column
    null_counts = df.isnull().sum()
    
    return (num_rows_with_nulls, null_counts)


def plot_correlation(df, column):
    # Drop rows with missing values in the specified column
    df = df.dropna(subset=[column])
    
    # Calculate the correlation coefficients
    corr = df.corr()[[column]]
    
    # Create a bar plot
    ax = sns.barplot(data=corr, x=column, y=corr.index, orient='h')
    ax.set_title(f'Correlation between {column} and other columns')
    
    # Show the plot
    plt.show()
```

```python
# Check for duplicates and null values in columns
data = remove_duplicates(df=data)
count_nulls(data)
```

```python
def find_outliers(df):
    outliers = {}
    fig, axs = plt.subplots(nrows=len(df.columns), ncols=1, figsize=(10, 5*len(df.columns)))
    
    # Loop through each column in the DataFrame
    for i, col in enumerate(df.columns):
        # Check if the column is numeric
        if df[col].dtype in ["int64", "float64"]:
            # Calculate the IQR
            Q1 = df[col].quantile(0.25)
            Q3 = df[col].quantile(0.75)
            IQR = Q3 - Q1
            
            # Calculate the lower and upper bounds
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR
            
            # Find the outlier indices
            outlier_indexes = df[(df[col] < lower_bound) | (df[col] > upper_bound)].index.tolist()
            
            # Add the outlier information to the dictionary
            outliers[col] = {
                'lower_bound': lower_bound,
                'upper_bound': upper_bound,
                'outlier_indexes': outlier_indexes,
                'outlier_values': df.loc[outlier_indexes, col].tolist()
            }
            
            # Plot the boxplot
            axs[i].boxplot(df[col], vert=False)
            axs[i].set_title(col)
    
    plt.show()
    return outliers
```

```python
outliers = find_outliers(data.drop(columns=columns_to_encode))
for col, info in outliers.items():
    print(f'Column: {col}')
    print(f'Outlier range: < {info["lower_bound"]} or > {info["upper_bound"]}')
    # print(f'Outlier rows: {info["outlier_indexes"]}')
    print(f'Outlier values: {info["outlier_values"]}')
```

```python
def plot_histograms(df, ncols=4):
    # Calculate the number of rows to use in the layout of the plots
    nrows = (len(df.columns) + ncols - 1) // ncols
    
    # Create a figure and axes
    fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(6*ncols, 4*nrows))
    
    # Flatten the axes array
    axes = axes.flatten()
    
    # Create a histogram for each column
    for i, column in enumerate(df.columns):
        ax = axes[i]
        sns.distplot(df[column].dropna(), ax=ax, hist=True, kde=True)
        #ax.set_title(column)
    
    # Show the plots
    plt.show()
```

```python
def min_max_scale(df):
   
    scaler = MinMaxScaler()
    
    # Fit the scaler to the data
    scaler.fit(df)
    
    # Transform the data
    scaled_data = scaler.transform(df)
    
    
    scaled_df = pd.DataFrame(scaled_data, columns=df.columns)
    
    return scaled_df

def standard_scale(df):
   
    scaler = StandardScaler()
    
    # Fit the scaler to the data
    scaler.fit(df)
    
    # Transform the data
    scaled_data = scaler.transform(df)
    
    
    scaled_df = pd.DataFrame(scaled_data, columns=df.columns)
    
    return scaled_df
```

```python
min_max_scaled_data = min_max_scale(data)
standard_scaled_data = standard_scale(data)
```

# K-means Clustering

```python
def kmeans_elbow_method(data, max_k=30, figsize=(15, 8)):
    ssd = []
    silhouette_scores = []
    plt.figure(figsize=figsize)
    # Loop through different values of k
    for k in range(2, max_k+1):
        kmeans = KMeans(n_clusters=k)
        
        # Fit the k-means clustering model to the data
        kmeans.fit(data)
        
        # Calculate the sum of squared distances
        ssd.append(kmeans.inertia_)
        
        # Calculate the average silhouette score
        silhouette_scores.append(silhouette_score(data, kmeans.labels_))
    
    fig, ax1 = plt.subplots()
    ax1.set_xlabel('Number of clusters (k)')
    ax1.set_ylabel('Sum of squared distances', color='b')
    ax1.plot(range(2, max_k+1), ssd, 'bo-')
    
    ax2 = ax1.twinx()
    ax2.set_ylabel('Silhouette score', color='r')
    ax2.plot(range(2, max_k+1), silhouette_scores, 'ro-')
    
    fig.tight_layout()
    plt.show()
```

```python
kmeans_elbow_method(standard_scaled_data, max_k=10)
```

```python
# Number of cluster for K-means to be 3
from sklearn.decomposition import PCA
k = 3
kmeans = KMeans(n_clusters = k)
kmeans.fit(standard_scaled_data)
cluster_labels = kmeans.labels_
kmeans_silhouette = silhouette_score(standard_scaled_data, cluster_labels)
dist = 1 - cosine_similarity(standard_scaled_data)
pca = PCA(2)
pca.fit(dist)
X_PCA = pca.transform(dist)

# Visualizing all the clusters 
plt.figure(figsize=(15,8))
sns.scatterplot(x=X_PCA[:, 0], y=X_PCA[:, 1], 
                hue=cluster_labels, palette=sns.color_palette('hls', k), s=50)
plt.title('Cluster of Customers', size=15, pad=10)
plt.legend(loc=0, bbox_to_anchor=[1,1])
plt.show()
```

```python
# Intepreting the clusters
from pandas.plotting import parallel_coordinates
kmeans_data = data.copy()
kmeans_data['kmeans_cluster'] = kmeans.labels_
kmeans_scaled_data = standard_scaled_data.copy()
kmeans_scaled_data['kmeans_cluster'] = kmeans.labels_
# Create the Parallel Coordinates Plot
plt.figure(figsize=(15,8))
parallel_coordinates(kmeans_scaled_data , 'kmeans_cluster', colormap='viridis')
# Rotate the x-axis labels
plt.xticks(rotation=90)
plt.show()
```

# DBSCAN

```python
def find_best_eps(data, min_samples):
    # Fit a nearest neighbors model to the data
    nn = NearestNeighbors(n_neighbors=min_samples)
    nn.fit(data)
    
    # Calculate the distances to the k-th nearest neighbor
    distances, _ = nn.kneighbors(data)
    distances = distances[:, -1]
    
    # Sort the distances in descending order
    distances = np.sort(distances)[::-1]
    
    # Plot the k-distance graph
    plt.plot(distances)
    plt.xlabel('Data point index')
    plt.ylabel(f'Distance to {min_samples}-th nearest neighbor')
    plt.show()
```

```python
find_best_eps(standard_scaled_data, min_samples=2)
```

```python
# Best epsilon is 1.4?, I am not sure about min_samples so I am going to use the number of columns
dbscan = DBSCAN(eps=1.5, min_samples=10)
dbscan.fit(standard_scaled_data)
cluster_labels = dbscan.labels_
dbscan_silhouette = silhouette_score(standard_scaled_data, cluster_labels)

dist = 1 - cosine_similarity(standard_scaled_data)
pca = PCA(2)
pca.fit(dist)
X_PCA = pca.transform(dist)

# Visualizing all the clusters 
plt.figure(figsize=(15,8))
sns.scatterplot(x=X_PCA[:, 0], y=X_PCA[:, 1], 
                hue=cluster_labels, palette=sns.color_palette('hls', k), s=50)
plt.title('Cluster of Customers', size=15, pad=10)
plt.legend(loc=0, bbox_to_anchor=[1,1])
plt.show()
```

# Ablation study 

```python
feature_names = standard_scaled_data.columns
# Initialize a dictionary to hold silhouette scores for each feature removed
kmeans_silhouette_scores = {}

# For each column in the credit card dataset
for i in range(standard_scaled_data.shape[1]):
    # Remove the column 
    column_to_drop = standard_scaled_data.columns[i]
    standard_scaled_data_new = standard_scaled_data.drop(column_to_drop, axis=1)
        
    # Run K-means clustering on the altered dataset
    kmeans = KMeans(n_clusters=k).fit(standard_scaled_data_new)

    # Calculate the silhouette score
    score = silhouette_score(standard_scaled_data_new, kmeans.labels_)

    # Store the silhouette score in the dictionary
    kmeans_silhouette_scores[feature_names[i]] = score

# Print the silhouette scores for each feature removed
for feature, score in kmeans_silhouette_scores.items():
    print(f"K-means Silhouette Score with {feature} removed: {score}")
```
