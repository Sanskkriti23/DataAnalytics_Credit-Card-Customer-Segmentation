```python
#Data preparation
data = data.fillna(0)
data = data.drop(["CUST_ID"], axis = 1)
data.head(5)
```

### Initial correlation matrix

```python
## file:///C:/Users/TANJ0292/Downloads/Credit_Card_Customer_Segmentation.pdf
corr_mat = data.corr()
for x in range(len(corr_mat)):
    corr_mat.iloc[x,x]=0
```

```python
## file:///C:/Users/TANJ0292/Downloads/Credit_Card_Customer_Segmentation.pdf

corr_mat.abs().max().sort_values(ascending=False)
```

```python
skew_cols = data.skew().sort_values(ascending=False)
skew_cols = skew_cols.loc[skew_cols > 0.75]
print(skew_cols)
```

```python
for col in skew_cols.index.tolist():
    data[col] = np.log1p(data[col])
```

```python
s = StandardScaler()
data = s.fit_transform(data)
data = pd.DataFrame(data, columns=corr_mat.columns)
data

#We can see how the data is now on the same scale. 
```

```python
## http://localhost:8888/notebooks/Downloads/Clustering.ipynb

from sklearn.decomposition import PCA

pca_list = list()
feature_weight_list = list()

# Fit a range of PCA models

for n in range(1, 18):
    
    # Create and fit the model
    PCAmod = PCA(n_components=n)
    PCAmod.fit(data)
    
    # Store the model and variance
    pca_list.append(pd.Series({'n':n, 'model':PCAmod, 'var': PCAmod.explained_variance_ratio_.sum()}))
    
    # Calculate and store feature importances
    abs_feature_values = np.abs(PCAmod.components_).sum(axis=0)
    feature_weight_list.append(pd.DataFrame({'n':n, 'features': data.columns, 'values':abs_feature_values/abs_feature_values.sum()}))
    
pca_df = pd.concat(pca_list, axis=1).T.set_index('n')
pca_df
```

```python
##file:///C:/Users/TANJ0292/Downloads/Credit_Card_Customer_Segmentation.pdf
PCAmoduse = PCA(n_components=11)
data_new = PCAmoduse.fit_transform(data)

data_new = pd.DataFrame(data_new, columns = ['col' + str(x) for x in range(1,12)])
data_new
```

```python
data_numpy = data_new.to_numpy()
data_numpy
```

```python
##Clustering with Affinity Propagation.
##This technique is very important as it segments the data without the need to specify the number of clusters.
preferences = np.arange(-30,-200,-10)
clusters = []

for p in preferences:
    af = AffinityPropagation(preference=p, damping=0.6, max_iter=400, verbose=False).fit(data_numpy)
    labels_af = pd.DataFrame(af.labels_)
    clusters.append(len(af.cluster_centers_indices_))

plt.figure(figsize=(10,7))
plt.xlabel("Preference")
plt.ylabel("Number of clusters")
plt.plot(preferences,clusters, marker='o')

##Insight: As the preferences parameter goes down so does the number of clusters.
##Insight: The minimum is attained for 5 clusters at -160. As such I will check the 5 clusters option
```

```python
# Compute Affinity Propagation
af = AffinityPropagation(preference =-160).fit(data_numpy)
cluster_centers_indices = af.cluster_centers_indices_
labels = af.labels_
  
n_clusters_ = len(cluster_centers_indices)





# Plot result
import matplotlib.pyplot as plt
from itertools import cycle
  
plt.close('all')
plt.figure(1)
plt.clf()
  
colors = cycle('bgrcmykbgrcmykbgrcmykbgrcmyk')
  
for k, col in zip(range(n_clusters_), colors):
    class_members = labels == k
    cluster_center = data_numpy[cluster_centers_indices[k]]
    plt.plot(data_numpy[class_members, 0], data_numpy[class_members, 1], col + '.')
    plt.plot(cluster_center[0], cluster_center[1], 'o',
             markerfacecolor = col, markeredgecolor ='k',
             markersize = 14)
  
    for x in data_numpy[class_members]:
        plt.plot([cluster_center[0], x[0]], 
                 [cluster_center[1], x[1]], col)
  
plt.title('Estimated number of clusters: % d' % n_clusters_)
plt.show()
```
